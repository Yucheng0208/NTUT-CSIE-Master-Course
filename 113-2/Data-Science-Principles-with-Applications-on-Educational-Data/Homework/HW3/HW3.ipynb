```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from the file path
exam = pd.read_csv('exam_results.csv')
```

```python
grouped_data = exam.groupby(['year', 'grade'])['society_2'].mean().unstack()

sns.heatmap(grouped_data, annot=True, cbar_kws={'label': 'Average Score'})
plt.title('Average Score by Level and Year')
plt.xlabel('Grade')
plt.ylabel('Year')
plt.show()
```

```python
# Write Answer here

selected_columns = ['chinese_1', 'math_1', 'english_1', 'science_1']

corr_matrix = exam[selected_columns].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap of Main Subjects')
plt.show()
```

### Discussion: Correlation Heatmap of Main Subjects  
The heatmap illustrates the correlation coefficients among the four main subjects: Chinese, Math, English, and Science. The results show moderate to strong positive correlations between all subjects, especially between Chinese and Science (0.72), and Math and Science (0.71). This suggests that students who perform well in one subject tend to do well in others. Such correlations help educators provide holistic support to students struggling in multiple subjects, rather than focusing on a single subject. For data analysts, understanding these relationships aids in predicting overall academic performance and designing more effective teaching strategies.
